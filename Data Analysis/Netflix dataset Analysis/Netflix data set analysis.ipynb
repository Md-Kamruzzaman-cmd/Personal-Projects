#### Q2. In which year, highest number of TV shows & Movies were released?

```python
data.dtypes
```

```python
data['release_date'] = pd.to_datetime(data['release_year'], format='%Y', errors='coerce')
```

```python
data.dtypes
```

```python
data.head()
```

```python
data['release_date'].dt.year.value_counts()
```

#### In 2018, highest number of movie or tv shows were released.

```python
plt.figure(figsize=(10, 4))
data['release_date'].dt.year.value_counts().nlargest(20).plot(kind='bar')
plt.xlabel('Year',fontsize = 12)
plt.ylabel('Occurrences')
plt.xticks(rotation=90)  # Rotate x-axis labels
plt.show()
```
